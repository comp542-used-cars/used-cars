```python
import pandas as pd
import re
import numpy as np
import sklearn as sk
```

```python
#df_vehicles = pd.read_csv("C:\\Users\\keith\\My Drive\\School\\CSUN\\COMP542\\Projects\\Data\\Raw\\vehicles.csv",low_memory=False)
df_usedcars = pd.read_csv("C:\\Users\\keith\\My Drive\\School\\CSUN\\COMP542\\Projects\\Data\\Raw\\used_cars_data.csv",low_memory=False)
```

```python
df_usedcars_limit = df_usedcars[['price', 'engine_cylinders','engine_displacement','city','engine_type','fuel_tank_volume','fuel_type','horsepower','is_new','latitude','longitude','height','length','width','make_name','model_name','mileage','maximum_seating','transmission','wheel_system','year']]
df_usedcars_limit = df_usedcars_limit.dropna(0,'any')
df_usedcars_limit
```

```python
df_clean_usedcars = pd.DataFrame()
df_clean_usedcars = df_usedcars_limit
df_normalized = pd.DataFrame()
df_normalized['price'] = df_clean_usedcars['price']

```

```python
#Raw Number Cols
cols = ['horsepower','latitude','mileage','year']
for col in cols:
    df_normalized[col] = (df_clean_usedcars[col]/(df_clean_usedcars[col].max()/2)) - 1   
```

```python
#True/False Cols
cols = ['is_new']
for col in cols:
    df_clean_usedcars[col] = df_clean_usedcars[col].replace(True,1).replace(False,0)
```

```python
#Clean columns that need alphas removed:
cols = ['engine_cylinders','engine_displacement','fuel_tank_volume','height','length','width','maximum_seating']

for col in cols:
    df_clean_usedcars[col] = pd.to_numeric(df_usedcars_limit[col].replace("[^0-9.]", "",regex=True).replace(r'^\s*$',None,regex=True),errors='ignore')
    df_normalized[col] = (df_clean_usedcars[col]/(df_clean_usedcars[col].max()/2)) - 1

 
   
df_normalized
```

```python
df_clean_usedcars['wheel_system'].unique()

#Type of Drive
for val in ['AWD','4WD','4X2']:  
    df_clean_usedcars['wheel_system'] = df_clean_usedcars['wheel_system'].replace(val,4)
for val in ['RWD','FWD']:  
    df_clean_usedcars['wheel_system'] = df_clean_usedcars['wheel_system'].replace(val,2)
    
df_normalized['wheel_system'] = (df_clean_usedcars['wheel_system']/(df_clean_usedcars['wheel_system'].max()/2)) - 1
```

```python
from sklearn.preprocessing import OneHotEncoder

cols = ['model_name','make_name']

enc = OneHotEncoder(handle_unknown='ignore')
for col in cols:
    transformed = enc.fit_transform(df_usedcars_limit[[col]])
    df_normalized[col] = transformed.toarray()
```

```python
df_normalized = df_normalized.dropna(0,'any')
df_normalized.astype({'price': 'float64'}).dtypes
df_normalized.info()

#Output Data
import io
folder = "C:\\Users\\keith\\My Drive\\School\\CSUN\\COMP542\\Projects\\Data\\tables"
name = "normalized_usedcars"
df_normalized.to_csv(f"{folder}/{name}.csv") 
df_normalized.describe(include='all').to_csv(f"{folder}/{name}_describe.csv")
df_normalized.corr(method='pearson',numeric_only=True).to_csv(f"{folder}/{name}_correlation.csv")
df_normalized.skew().to_csv(f"{folder}/{name}_skew.csv")
buffer = io.StringIO()
df_normalized.info(verbose=True,buf=buffer)
s = buffer.getvalue()
with open(f"{folder}/{name}_info.csv","w+") as f:
    f.write(s)
    
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error
df_cut = df_usedcars_limit[['price','horsepower','year','mileage']].dropna(0,'any').sample(n=100000)
X = df_cut.drop('price',axis=1)
y = df_cut['price']
y=y.astype('int64')
X_train, X_test, y_train, y_test = train_test_split(X, y)
y_train=y_train.astype('int64')
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
X.shape
X_new = SelectKBest(chi2, k=5).fit_transform(X, y)
X_new.shape
```

```python
from sklearn.linear_model import BayesianRidge
BayReg = BayesianRidge()
BayReg = BayReg.fit(X_train, y_train)
y_pred = BayReg.predict(X_test)
print(mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.linear_model import LinearRegression
olr = LinearRegression()
olr = olr.fit(X_train, y_train)
y_pred = olr.predict(X_test)
print(mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.linear_model import ARDRegression, LinearRegression, BayesianRidge

olr = LinearRegression().fit(X, y)
brr = BayesianRidge(compute_score=True, n_iter=30).fit(X, y)
ard = ARDRegression(compute_score=True, n_iter=30).fit(X, y)
linear_df = pd.DataFrame(
    {
        "ARDRegression": ard.coef_,
        "BayesianRidge": brr.coef_,
        "LinearRegression": olr.coef_,
    }
)

linear_df
```

```python
from sklearn import tree
clf = tree.DecisionTreeRegressor()
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(mean_absolute_percentage_error(y_test,y_pred))
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#33.3% error

from sklearn.naive_bayes import GaussianNB
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
y_test=y_test.astype('int64')
y_train=y_train.astype('int64')
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
print(mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                     hidden_layer_sizes=(15,), random_state=1)

clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(mean_absolute_percentage_error(y_test,y_pred))
```
